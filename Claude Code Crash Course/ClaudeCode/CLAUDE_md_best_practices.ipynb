# 🧠 CLAUDE.md Best Practices Notebook

A comprehensive, practical, and **production-grade guide** for designing `CLAUDE.md` files.

This notebook is based on official Anthropic guidance and real-world best practices.

---

## 📌 Key Concepts

- CLAUDE.md = **Persistent Instructions Layer**
- Auto Memory = **Learned Behavior Layer**
- Rules / agent_docs = **On-demand Context Layer**


## 🧩 System Architecture

```mermaid
flowchart TD
    A[User Prompt] --> B[Claude System Prompt]
    B --> C[CLAUDE.md]
    C --> D[Auto Memory]
    D --> E[Rules / agent_docs]
    E --> F[Final Response]
```

**Insight:** CLAUDE.md is **not enforced**, only *influential*.

## ⚠️ Critical Reality

- Claude **may ignore CLAUDE.md** if irrelevant
- Instructions compete for attention
- More instructions = worse adherence

```mermaid
graph LR
    A[Few Instructions] --> B[High Accuracy]
    C[Too Many Instructions] --> D[Low Accuracy]
```


## 🧱 WHAT / WHY / HOW Framework

```mermaid
flowchart TD
    A[CLAUDE.md]
    A --> B[WHAT: Tech + Structure]
    A --> C[WHY: Purpose]
    A --> D[HOW: Workflows]
```

### Example

- WHAT → Next.js + Prisma
- WHY → SaaS dashboard
- HOW → `pnpm dev`, `pnpm test`


## 📦 Loading Mechanism

```mermaid
flowchart TD
    A[Root CLAUDE.md]
    A --> B[Parent Directory CLAUDE.md]
    B --> C[Subdirectory CLAUDE.md]
    C --> D[Rules Triggered]
```

- All are **concatenated**, not overridden
- Local overrides come last


## 📁 Progressive Disclosure Pattern

```mermaid
flowchart TD
    A[CLAUDE.md - Minimal]
    A --> B[agent_docs]
    B --> C[Deep Context Files]
    C --> D[Only Loaded When Needed]
```

### Folder Example

```
agent_docs/
  ├── testing.md
  ├── architecture.md
  ├── database.md
```


## 📏 Size & Structure Rules

- Ideal: **< 200 lines**
- Use **bullets, headers**
- Avoid long paragraphs

```mermaid
graph TD
    A[Short File] --> B[High Compliance]
    C[Long File] --> D[Ignored Instructions]
```


## 🚫 Anti-Patterns

```mermaid
flowchart TD
    A[Bad CLAUDE.md]
    A --> B[Too Many Rules]
    A --> C[Irrelevant Info]
    A --> D[Style Guides]
    A --> E[Huge File]
```

- Don't use as linter
- Don't dump everything
- Don't auto-generate blindly


## 🧠 Auto Memory vs CLAUDE.md

```mermaid
flowchart LR
    A[CLAUDE.md] --> B[User Written]
    C[Auto Memory] --> D[Claude Learned]
```

- CLAUDE.md = rules
- Auto memory = learnings


## 🏗️ Ideal CLAUDE.md Template

```markdown
# Project Overview
- What system does

# Tech Stack
- Frameworks, DB, tools

# Project Structure
- Key folders

# Workflows
- Build, test, run

# Rules
- Critical always-follow rules

# External Docs
- @agent_docs/testing.md
```
